```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("out", "r") as ins:
    array = []
    for line in ins:
        array.append(line)
```

```python
def get_values(array, prefix):

    insert = [line for line in array if line.startswith(prefix)]
    val = np.full((len(insert),2), np.nan)

    for i,e in enumerate(insert):
        tmp = e.split(",")
        try:
            val[i, 0] = tmp[1]
            val[i, 1] = tmp[2]
        except:
            continue
        
    return val

def moving_average(a, window_size):
    
    ret = np.full(a.shape, np.nan)
    
    for i in range(int(len(val) / window_size)):
        _from = i * window_size
        _until = min(_from + window_size, len(val))
        I = np.arange(_from, _until)
        ret[I] = a[I].mean()
    
    return ret
    

```

```python


val = get_values(array, "[INSERT]")[10:]
window_size = int(len(val) / 3)

avg = moving_average(val[:, 1], window_size)

plt.plot(val[:, 0], val[:,1], alpha=0.7)
plt.scatter(val[:, 0], val[:,1], alpha=0.3)

plt.plot(val[:, 0], avg, alpha=1.0, color="red")

plt.title("INSERT")
plt.xlabel("Time (ms)")
plt.ylabel("Update Visibility Latency (ms)")


plt.savefig("insert.pdf")
#plt.show()
```
